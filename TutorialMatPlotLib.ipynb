# **Matplotlib - Interfaz orientada a objetos**

Si bien es fácil generar gráficos rápidamente con el módulo *matplotlib.pyplot*, se recomienda el uso del enfoque orientado a objetos, ya que brinda más control y personalizadción de sus gráficos, La mayoría de las funciones también están disponibles en la clase *matplotlib.axes.Axes*.
La idea principal detrás del uso del método orientado a objetos más formal es crear objetos de figura y luego simplemente llamar a métodos o atributos fuera de ese objeto.
Este enfoque ayuda a manejar mejor un lienzo que tiene varias tramas.

En la Interfaz orientada a objetos, Pyplot se usa solo para algunas funciones, como la creación de figuras, y el usuario crea y realiza un seguimiento explícito de los objetos de figura y ejes, se pueden crear uno o más objetos de ejes. # **Matplotlib - Funcion Subplots ()**

La API spyplot de Matplotlib tiene una función conveniente llamada subplots() que actúa como un contenedor de utilidades y ayuda a crear diseños comunes de subtramas, incluido el objeto de figura adjunto, en una sola llamada



```
plt.subplots(nrows, ncols)
```

Los dos argunmentos enteros de esta función especifican el número de filas y columnas de la cuadrícula de la subparcela. La función devuelve un objeto de figura y una tupla que contiene objetos de ejes iguales a nrows*ncols. Cada objeto de ejes es accesible por su índice. Aquí creamos una subparcela de 2 filas por 2 columnas y mostramos 4 parcelas diferentes en cada subparcela.

````python
import matplotlib.pyplot as plt
fig,a = plt.subplots(2,2)
import numpy as np
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('Cuadrado')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('Raiz Cuadrada')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('Exponencial')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('Logaritmo')
plt.show()
````

# **Matplotlib - Función Subplot2grid()**

Esta función brinda más flexibilidad para crear un objeto de ejes en una ubicación especifica de la cuadrícula. También permite que el objeto de los ejes se extienda por varias filas o columnas

```
Plt.subplot2grid(shape, location, rowspan, colspan)
```

En el siguiente ejemplo, una cuadrícula de 3X3 del objeto de figura se llena con objetos de ejes de diferentes tamaños en rangos de filas y columnas, cada uno mostrando una gráfica diferente

````python
import matplotlib.pyplot as plt
a1 = plt.subplot2grid((3,3),(0,0), colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0), rowspan = 2, colspan = 2)
import numpy as np
x = np.arange(1,10)
a2.plot(x, x*x)
a2.set_title('Cuadrado')
a1.plot(x, np.exp(x))
a1.set_title('Exponencial')
a3.plot(x, np.log(x))
a3.set_title('Logaritmo')
plt.tight_layout()
plt.show()
````

# **Matplotlib - Cuadrículas**

La función *gird()* del objeto de ejes activa o desactiva la visibilidad de la cuadrícula dentro de la figura, También puede mostrar los ticks principales/menores (o ambos) de la cuadrícula. Además, las propiedades de color, estilo de línea y ancho de línea se pueden configurar en la función grid()

````python
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(1, 3, figsize = (12, 4))
x = np.arange(1, 11)
axes[0].plot(x, x**3, 'g', lw = 2)
axes[0].grid (True)
axes[0].set_title('Cuadrícula Predeterminada')
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color = 'b', ls = '-.', lw = 0.25)
axes[1].set_title('Cuadrícula Personalizada')
axes[2].plot(x, x)
axes[2].set_title('Sin Cuadrícula')
fig.tight_layout()
plt.show()
````

# **Matplotlib - Ejes de Formato**

A veces, uno o unos pocos puntos son mucho más grandes que la mayor parte de los datos. En tal caso, la escala de un eje debe establecerse como logaritmica en lugar de la escala normal. Esta es la escala logarítmica. En Matplotlib, es posible estableces la propiedad xscale o xscale del objeto de ejes en 'registro'.

A veces también se requiere mostrar alguna distancia adicional entre los números de eje y la etiqueta del eje. La propiedad del rotulador de cualquiera de los ejes (*x* o *y* o ambos) se puede estableces en el valor deseado.

Ambas características anteriores se demuestran con la ayuda del siguiente ejemplo. La subparcela de la derecha tiene una escala logarítmica y la de la izquierda tiene su eje *x* con una etiqueta a más distancia.

````python
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(1, 2, figsize = (10, 4))
x = np.arange(1, 5)
axes[0].plot(x, np.exp(x))
axes[0].plot(x, x**2)
axes[0].set_title("Escala Normal")
axes[1].plot (x, np.exp(x))
axes[1].plot (x, x**2)
axes[1].set_yscale("log")
axes[1].set_title("Escala Logaritmica (y)")
axes[0].set_xlabel("Eje x")
axes[0].set_ylabel("Eje y")
axes[0].xaxis.labelpad = 10
axes[1].set_xlabel("Eje x")
axes[1].set_ylabel("Eje y")
plt.show()
````

Las espinas de los ejes son las líneas que conectan las marcas de los ejes que delimitan los límites del área de la parcela. El objeto de ejes tiene espinas ubicadas en la parte superios, inferior, izquierda y derecha.

Cada lomo se puede formatear especificando el color y el ancho. Cualquier borde se puede hacer invisible si su color se establece en ninguno.

````python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color(None)
ax.spines['top'].set_color(None)
ax.plot([1, 2, 3, 4, 5])
plt.show()
````

Un diagrama de barras o gráfico de barras es un cuadro o gráfico que presenta datos categóricos con barras restangulares con alturas o longitudes proporcionales a los valores que representan. Las barras se pueden trazar vartical u horizontalmente.

Un gráfico de barras muestra comparaciones entre categorías discretas. Un eje del gráfico muestra las categorias especificas que se comparan y el otro eje representa un valor medido.

La API de Matplotlib proporciona la función *bar()* que se puede usar en el estilo de MATLAB, así como en la API orientada a objetos. La firma de la función *bar()* que se utilizará con el objeto de ejes es la siguiente:

```
ax.bar(x, height, width, bottom, align)
```

La función crea un gráfico de barras con el restángulo delimitado de tamaño (x - ancho = 2; x + ancho = 2; fondo; fondo + alto).

  - x : secuencia de escalares que representan las coordenadas *x* de las barras. alinnee los controles si *x* es el centro de la barra (predeterminado) o el borde izquierdo.

  - altura: escalar o secuencia de escalares que representan la(s) altura(s) de las barras.

  - ancho: escalar o tipo matriz, opcional. el(los) ancho(s) de las barras por defecto 0.8

  - abajo: escalar o tipo matriz, opcional. la(s) coordenada(s) y de las barras por defecto es Ninguna.

  - alinear: {'center', 'edge'}, opcional, predeterminado 'center'

La función devuelve un objeto contenedor Matplotlib con todas las barras.

#**Matplotlib - Histograma**
Un histograma es una representación precisa de la distribución de datos numéricos. Es una estimación de la distribución de probabilidad de una variable continua. Es una especie de gráfico de barras.

Para construir un histograma, siga estos pasos:

  - Bin el rango de valores.
  - Divide todo el rango de valores en una serie de intervalos.
  - Cuente cuántos valores caen en cada intervalo.

Los contenedores generalmente se especifican como intervalos consecutivos que no se superponen de una variable.

La función matplotlib.pyplot.hist() traza un histograma. Calcula y dibuja el histograma de x.

**Parámetros**

La siguiente tabla enumera los parámetros para un histograma:

  - X: array o secuencia de arrays
  - contenedores:	entero o secuencia o 'auto', opcional

parámetros opcionales:

  - rango:	El rango inferior y superior de los contenedores.
  - densidad:	Si es Verdadero, el primer elemento de la tupla de retorno serán los conteos normalizados para formar una densidad de probabilidad
  - acumulativo:	Si es Verdadero, se calcula un histograma en el que cada contenedor proporciona los recuentos en ese contenedor más todos los contenedores para valores más pequeños.
  - tipo de hist:	El tipo de histograma a dibujar. El valor predeterminado es 'barra'
 - 'bar' es un histograma tradicional de tipo barra. Si se dan varios datos, las barras se organizan una al lado de la otra.
 - 'barstacked' es un histograma de tipo barra en el que se apilan varios datos uno encima del otro.
 - 'paso' genera un diagrama de líneas que por defecto está vacío.
 - 'stepfilled' genera un gráfico de líneas que se rellena de forma predeterminada.

El siguiente ejemplo traza un histograma de calificaciones obtenidas por los estudiantes en una clase. Se definen cuatro contenedores, 0-25, 26-50, 51-75 y 76-100. El histograma muestra el número de estudiantes que se encuentran en este rango.

````python
from matplotlib import pyplot as plt
import numpy as np
fig,ax = plt.subplots(1,1)
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
ax.hist(a, bins = [0,25,50,75,100])
ax.set_title("histogram of result")
ax.set_xticks([0,25,50,75,100])
ax.set_xlabel('marks')
ax.set_ylabel('no. of students')
plt.show()
````

#**Matplotlib - Gráfico circular**

Un gráfico circular solo puede mostrar una serie de datos. Los gráficos circulares muestran el tamaño de los elementos (llamado cuña) en una serie de datos, proporcional a la suma de los elementos. Los puntos de datos en un gráfico circular se muestran como un porcentaje de todo el gráfico circular.

La API de Matplotlib tiene una función pie() que genera un diagrama circular que representa los datos en una matriz. El área fraccionaria de cada cuña viene dada por x/sum(x) . Si sum(x)< 1, entonces los valores de x dan directamente el área fraccionaria y la matriz no se normalizará. El pastel resultante tendrá una cuña vacía de tamaño 1 - sum(x).

El gráfico circular se ve mejor si la figura y los ejes son cuadrados, o si el aspecto de los ejes es igual.

**Parámetros**

La siguiente tabla enumera los parámetros de un gráfico circular:
 - X:	tipo matriz. Los tamaños de cuña.

 - Colores:	Una secuencia de matplotlibcolorargs a través de la cual se desplazará el gráfico circular. Si es Ninguno, usará los colores en el ciclo actualmente activo.
 - Autopcto:	cadena, utilizada para etiquetar las cuñas con su valor numérico. La etiqueta se colocará dentro de la cuña. La cadena de formato será fmt%pct.

El siguiente código utiliza la función pie() para mostrar el gráfico circular de la lista de estudiantes inscritos en varios cursos de lenguaje informático. El porcentaje proporcional se muestra dentro de la cuña respectiva con la ayuda del parámetro autopct que se establece en %1.2f%.

# **Matplotlib - Gráfico de contorno**

Los gráficos de contorno (a veces llamados gráficos de nivel) son una forma de mostrar una superficie tridimensional en un plano bidimensional. Grafica dos variables predictoras XY en el eje y y una variable de respuesta Z como contornos. Estos contornos a veces se denominan cortes z o valores de respuesta iso.

Una gráfica de contorno es adecuada si desea ver cómo cambia el valor Z en función de dos entradas X e Y, de modo que Z = f(X,Y). Una línea de contorno o isolínea de una función de dos variables es una curva a lo largo de la cual la función tiene un valor constante.

Las variables independientes x e y generalmente están restringidas a una cuadrícula regular llamada meshgrid. El numpy.meshgrid crea una cuadrícula rectangular a partir de una matriz de valores x y una matriz de valores y.

La API de Matplotlib contiene funciones de contorno() y contornof() que dibujan líneas de contorno y contornos rellenos, respectivamente. Ambas funciones necesitan tres parámetros x, y y z.


````python
import numpy as np
import matplotlib.pyplot as plt
xlist = np.linspace(-3.0, 3.0, 100)
ylist = np.linspace(-3.0, 3.0, 100)
X, Y = np.meshgrid(xlist, ylist)
Z = np.sqrt(X**2 + Y**2)
fig,ax=plt.subplots(1,1)
cp = ax.contourf(X, Y, Z)
fig.colorbar(cp) # Add a colorbar to a plot
ax.set_title('Filled Contours Plot')
#ax.set_xlabel('x (cm)')
ax.set_ylabel('y (cm)')
plt.show()
````

# Matplotlib- Gráfico de contorno 3D

La función ax.contour3D() crea un gráfico de contorno tridimensional. Requiere que todos los datos de entrada estén en forma de cuadrículas regulares bidimensionales, con los datos Z evaluados en cada punto. Aquí, mostraremos un diagrama de contorno tridimensional de una función sinusoidal tridimensional.


````python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
def f( x , y):
  return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6 , 30)
y = np.linspace(-6, 6 , 30)

X , Y = np.meshgrid(x, y)
Z = f(X , Y)

fig = plt.figure()
ax = plt.axes (projection = '3d')
ax.contour3D (X , Y , Z ,50, cmap ='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Contorno 3D')
plt.show()
````

# Matplotlib - Diagrama de estructura alámbrica 3D
El diagrama de estructura alámbrica toma una cuadrícula de valores y la proyecta sobre la superficie tridimensional especificada, y puede hacer que las formas tridimensionales resultantes sean bastante fáciles de visualizar. La función plot_wireframe() se usa para este propósito:

````python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
def f( x , y):
  return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-6, 6 , 30)
y = np.linspace(-6, 6 , 30)

X , Y = np.meshgrid(x, y)
Z = f(X , Y)

fig = plt.figure()
ax = plt.axes (projection = '3d')
ax.plot_wireframe(X,Y,Z, color ='blue')
ax.set_title('Estructura Alámbrica')
plt.show()
````

# Matplotlib - Gráfico de superficie 3D
El gráfico de superficie muestra una relación funcional entre una variable dependiente designada (Y) y dos variables independientes (X y Z). El gráfico es un gráfico complementario del gráfico de contorno. Un gráfico de superficie es como un gráfico de estructura alámbrica, pero cada cara de la estructura alámbrica es un polígono relleno. Esto puede ayudar a la percepción de la topología de la superficie que se está visualizando. La función plot_surface() x, y y z como argumentos.

````python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
x = np.outer (np.linspace(-2, 2,30), np.ones(30))
y = x.copy().T #transpose
z = np.cos(x**2 + y**2)

fig = plt.figure()
ax = plt.axes (projection = '3d')

ax.plot_surface(x,y,z, cmap='viridis', edgecolor= 'none')
ax.set_title('Gráfico de Superficie')
plt.show()
````

# Matplotlib - Trabajar con imágenes
El módulo de imagen en el paquete Matplotlib proporciona las funcionalidades necesarias para cargar, cambiar la escala y mostrar la imagen.

La biblioteca de Pillow admite la carga de datos de imagen. De forma nativa, Matplotlib solo admite imágenes PNG. Los comandos que se muestran a continuación recurren a Pillow si falla la lectura nativa.

La imagen utilizada en este ejemplo es un archivo PNG, pero tenga en cuenta el requisito de Pillow para sus propios datos. La función imread() se usa para leer datos de imagen en un objeto ndarray de float32 dtype.

````python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
img = mpimg.imread('Escudo_de_la_Universidad_Nacional_de_Colombia_(2016).svg.png')
````

Cualquier matriz que contenga datos de imagen se puede guardar en un archivo de disco ejecutando la función imsave() . Aquí se guarda una versión volteada verticalmente del archivo png original dando el parámetro de origen como inferior.

````python
plt.imsave("logo.png", img, cmap = 'gray', origin = 'lower')
````
